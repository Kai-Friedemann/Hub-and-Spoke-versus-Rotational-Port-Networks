```python
import numpy as np
import matplotlib.pyplot as plt

# Define the map F: X -> X
def F(x):
    # Example: the logistic map
    r = 3.7
    return r * x * (1 - x)

# Parameters
x0 = 0.5            # Initial condition
n_steps = 100       # Number of iterations

# Generate the orbit
orbit = [x0]
x = x0
for _ in range(n_steps):
    x = F(x)
    orbit.append(x)

# Convert to numpy array for convenience
orbit = np.array(orbit)

# Plot the orbit over time
plt.figure(figsize=(8, 5))
plt.plot(range(n_steps + 1), orbit, marker='o', linestyle='-')
plt.title('Orbit over Time')
plt.xlabel('Time step $t$')
plt.ylabel('State $x_t$')
plt.grid(True)
plt.show()

# Plot phase portrait: (x_t, x_{t+1})
plt.figure(figsize=(6, 6))
plt.plot(orbit[:-1], orbit[1:], 'o-')
plt.title('Phase Portrait')
plt.xlabel('$x_t$')
plt.ylabel('$x_{t+1}$')
plt.grid(True)
# plt.savefig("quadratic_phase_portrait.pdf")
plt.show()
```
