```python
import numpy as np
import matplotlib.pyplot as plt

# Discontinuous forcing term
def B(x):
    return np.where(x <= 0.5, 0.5, -0.5)

# Right-hand side of the ODE
def dxdt(x):
    return x * (1 - x) + B(x)

# Euler method solver
def forward_euler(x0, h, T):
    steps = int(T / h)
    x = np.zeros(steps)
    x[0] = x0
    for n in range(steps - 1):
        x[n + 1] = x[n] + h * dxdt(x[n])
    return x

# Parameters
x0 = 0.6
h = 0.1
T = 20.0

# Run simulation
t = np.arange(0, T, h)
x = forward_euler(x0, h, T)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='System Trajectory', color='blue', linewidth=1)
plt.axhline(0.5, color='r', linestyle='--', label='Threshold (x = 0.5)')
plt.xlabel('Time', fontsize=12)
plt.ylabel('$x(t)$', fontsize=12)
plt.title('Discontinuous Forcing Dynamics with $x_0 = 0.6$', fontsize=14)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)

# Optional save
# plt.savefig("quadratic_map_with_binary_switch.pdf", format='pdf')

plt.show()
```
