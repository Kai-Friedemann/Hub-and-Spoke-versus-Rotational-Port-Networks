```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_map(x, mu):
    """Quadratic (logistic) map function"""
    return mu * x * (1 - x)

def generate_trajectory(mu, x0, n_iter=10):
    """Generate trajectory of the quadratic map"""
    trajectory = np.zeros(n_iter)
    trajectory[0] = x0
    for i in range(1, n_iter):
        trajectory[i] = quadratic_map(trajectory[i-1], mu)
    return trajectory

# Parameters
mu = 3.0       # Parameter value (note: >4 leads to escape to infinity)
x0 = 0.9       # Initial condition
n_iter = 15    # Number of iterations

# Generate trajectory
trajectory = generate_trajectory(mu, x0, n_iter)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(trajectory, 'b-o', markersize=4, label=f'μ = {mu}, x₀ = {x0}')

plt.title(f'Quadratic Map Trajectory (μ = {mu})', fontsize=14)
plt.xlabel('Iteration (n)', fontsize=12)
plt.ylabel('xₙ', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-.1, 1.1)
plt.savefig('quadratic_map_5_9.pdf',
            format='pdf',
            bbox_inches='tight',
            dpi=300)
plt.show()
# plt.savefig(fname= "quadratic_map_5_9.pdf", format='pdf')
```
